```python
def is_false_positive(pred, actual, index):
    if pred[index] == 1 and actual[index] != 'passed':
        #print('found false positive at {}'.format(i))
        return True
```

```python
def is_true_negative(pred, actual, index):
    if pred[index] == 0 and actual[index] != 'passed':
        #print('found true negative at {}'.format(i))
        return True
```

```python
def return_positive_streak(actual, first, last):
    return actual[first:last+1].count('passed')
```

```python
projects = ['gradle', 'cloud_controller_ng', 'geoserver', 'chef', 'jruby', 'opal', 'orbeon-forms']
for p in projects:
    print('processing ' + p + ' ...')
    mf = pd.read_csv(metrics_path + p + '_real_metrics.csv')
    tf = pd.read_csv(original_path + p + '_test.csv')
    
    pred_result = mf['Build_Result'].tolist()
    actual_result = tf['tr_status'].tolist()
    
    fp_index = []
    tn_index = []
    
    fp_flag = False
    tn_flag = False
    
    non_delay_streak = []
    delay_streak = []
    
    for i in range(len(pred_result)):
        #print((pred_result[i], actual_result[i]))
        fp_flag = False
        tn_flag = False
        if is_false_positive(pred_result, actual_result, i):
            fp_index.append(i)
            fp_flag = True
        
        if is_true_negative(pred_result, actual_result, i):
            
            if len(fp_index) == 0 and len(tn_index) == 0:
                tn_index.append(i)
                continue
            elif len(fp_index) > 0 and len(tn_index) == 0:
                fp_flag = True
            elif len(fp_index) == 0 and len(tn_index) > 0:
                tn_flag = True
            else:
                if len(fp_index) > 0 and len(tn_index) > 0:
                    if fp_index[-1] < tn_index[-1]:
                        tn_flag = True
                    else:
                        fp_flag = True
            
            if tn_flag == True:
                last = tn_index.pop()
                non_delay_streak.append(return_positive_streak(actual_result, last, i))
                #print(non_delay_streak)
            
            if fp_flag == True:
                last = fp_index.pop()
                delay_streak.append(return_positive_streak(actual_result, last, i))
                if len(tn_index) == 0:
                    for e in fp_index:
                        delay_streak.append(return_positive_streak(actual_result, e, i))
                else:
                    for e in fp_index:                        
                        if e > tn_index[-1]:
                            delay_streak.append(return_positive_streak(actual_result, e, i))

            tn_index.append(i)
    
    print(non_delay_streak)
    print(delay_streak)
```
