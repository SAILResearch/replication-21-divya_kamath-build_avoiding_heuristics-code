```python
import pandas as pd
from statistics import mean, median
```

```python
sbs_results = pd.read_csv('final_sbs_results.csv')
gr_results = pd.read_csv('final_grouping_results.csv')
mlci_results = pd.read_csv('final_mlci_results.csv')
tr_results = pd.read_csv('final_tr_results.csv')
tr_best_confidence = pd.read_csv('best_confidence.csv')
```

```python
gr_new_b4_results = gr_results[(gr_results['algorithm'] == 'BATCHSTOP4') & (gr_results['batch_size'] == 4)]
tr_new_b4_results = tr_results[(tr_results['algorithm'] == 'BATCHSTOP4') & (tr_results['batch_size'] == 4)]
mlci_new_b4_results = mlci_results[(mlci_results['algorithm'] == 'BATCHSTOP4') & (mlci_results['batch_size'] == 4)]
tr_bc_b4_results = tr_best_confidence[(tr_best_confidence['algorithm'] == 'BATCHSTOP4') & (tr_best_confidence['batch_size'] == 4)]
```

```python
tr_new_b4_results
```

```python
len(gr_new_b4_results), len(tr_new_b4_results), len(mlci_new_b4_results)
```

```python
gr_new_b4_results['algorithm'] = ['BATCH4']*len(gr_new_b4_results)
tr_new_b4_results['algorithm'] = ['BATCH4']*len(tr_new_b4_results)
mlci_new_b4_results['algorithm'] = ['BATCH4']*len(mlci_new_b4_results)
tr_bc_b4_results['algorithm'] = ['BATCH4']*len(tr_bc_b4_results)
```

```python
sbs_results = sbs_results[~((sbs_results['batch_size'] == 4) & (sbs_results['algorithm'] == 'BATCHSTOP4'))]
gr_results = gr_results[~((gr_results['batch_size'] == 4) & ((gr_results['algorithm'] == 'BATCH4') | (gr_results['algorithm'] == 'BATCHSTOP4')))]
```

```python
tr_results = tr_results[~((tr_results['batch_size'] == 4) & ((tr_results['algorithm'] == 'BATCH4') | (tr_results['algorithm'] == 'BATCHSTOP4')))]
tr_best_confidence = tr_best_confidence[~((tr_best_confidence['batch_size'] == 4) & ((tr_best_confidence['algorithm'] == 'BATCH4') | (tr_best_confidence['algorithm'] == 'BATCHSTOP4')))]
mlci_results = mlci_results[~((mlci_results['batch_size'] == 4) & ((mlci_results['algorithm'] == 'BATCH4') | (mlci_results['algorithm'] == 'BATCHSTOP4')))]
```

```python
gr_results = gr_results.append(gr_new_b4_results)
tr_results = tr_results.append(tr_new_b4_results)
mlci_results = mlci_results.append(mlci_new_b4_results)
tr_best_confidence = tr_best_confidence.append(tr_bc_b4_results)
```

```python
len(gr_results), len(tr_results), len(mlci_results), len(sbs_results)
```

```python
sbs_results = sbs_results.rename(columns={'length':'testall_size'})
```

```python
sbs_results['method'] = ['sbs']*len(sbs_results)
gr_results['method'] = ['gr']*len(gr_results)
mlci_results['method'] = ['mlci']*len(mlci_results)
tr_results['method'] = ['tr']*len(tr_results)
```

```python
print(sbs_results.columns, gr_results.columns, mlci_results.columns, tr_results.columns)
```

```python
dataframes = [sbs_results, gr_results, mlci_results, tr_results]
```

```python
plot_dfs = [sbs_results, gr_results, mlci_results, tr_best_confidence]
```

```python
result_df = pd.concat(dataframes)
```

```python
result_df.to_csv('final_results.csv')
```

```python
plot_dfs = pd.concat(plot_dfs)
```

```python
plot_dfs = plot_dfs.drop('ci', axis=1)
plot_dfs = plot_dfs.drop('delay_list', axis=1)
plot_dfs.to_csv('final_results_for_plotting.csv')
```
