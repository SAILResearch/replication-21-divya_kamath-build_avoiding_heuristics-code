```python
import pandas as pd
from statistics import mean, median
import sys
import pprint
import pickle
```

```python
def str_to_list(s):
    if s == '[]':
        return [0]
    l = s[1:-1].split(', ')
    l = [int(x) for x in l]
    return l
```

```python
list_file = pd.read_csv('short_combined_results.csv')
comp_file = pd.read_csv('full_combined_delay.csv')
```

```python
list_file = list_file.sort_values(by=['project', 'algorithm', 'batch_size', 'confidence'])
comp_file = comp_file.sort_values(by=['project', 'algorithm', 'batch_size', 'confidence'])
```

```python
comp = comp_file['zero_median'].tolist()
```

```python
for x in range(len(list_file)):
    row = list_file.iloc[x]
    
    p = str_to_list(row['project_delays'])
    b = str_to_list(row['batch_median'])
    total_delay =  p + b
    
    print(median(total_delay))
    print(comp[x])
    print('====')
```
