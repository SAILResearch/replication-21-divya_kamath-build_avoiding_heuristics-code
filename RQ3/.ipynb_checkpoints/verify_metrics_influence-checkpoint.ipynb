```python
import pickle
import matplotlib
```

```python
project_list = ['rails.csv', 'cloudify.csv', 'gradle.csv', 'sonarqube.csv', 'loomio.csv', 'fog.csv', 'opal.csv', 'cloud_controller_ng.csv', 'puppet.csv', 'concerto.csv', 'sufia.csv', 'orbeon-forms.csv', 'graylog2-server.csv']
```

```python
for project in project_list:
    pkl_file = 'pickled_data/rq3_' + project + '_aici_model.pkl'
    print(pkl_file)
    
    with open(pkl_file, 'rb') as load_file:
        model = pickle.load(load_file)
    
    important_features_dict = {}
    for idx, val in enumerate(model.feature_importances_):
        important_features_dict[idx] = val

    important_features_list = sorted(important_features_dict,
                                 key=important_features_dict.get,
                                 reverse=True)
    print(important_features_list)
        
    print('\n\n\n')
    
```
